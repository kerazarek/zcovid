```python
daily_csvs = sorted(daily_dir.glob("*.csv"), key=lambda f: f.name)
daily_dfs = list(map(pd.read_csv, daily_csvs))
daily_combined = daily_dfs[0]
daily_combined
```

```python
for df in daily_dfs[1:]:
    daily_combined = daily_combined.merge(df, 'outer')
```

```python
daily_combined
```

```python
daily_combined.columns = ['subregion', 'region', 't', 'confirmed', 'deaths', 'recovered', 'latitude', 'longitude']
daily_combined.t = pd.to_datetime(daily_combined.t)
# daily_combined.loc[daily_combined.loc[:, ('confirmed', 'deaths', 'recovered')].isna()] = 0

daily_combined.loc[daily_combined.confirmed.isna(), 'confirmed'] = 0
daily_combined.loc[daily_combined.deaths.isna(), 'deaths'] = 0
daily_combined.loc[daily_combined.recovered.isna(), 'recovered'] = 0

daily_combined.confirmed = daily_combined.confirmed.astype(int)
daily_combined.deaths = daily_combined.deaths.astype(int)
daily_combined.recovered = daily_combined.recovered.astype(int)
```

```python
# daily_combined.confirmed.describe()
```

```python
# subregions = list(zip(daily_combined.region, daily_combined.subregion))
regions = []
subregions = []
region_subregions = {}
for r, s in zip(daily_combined.region, daily_combined.subregion):
    if r not in regions:
        regions.append(r)
        region_subregions[r] = []
    if (r, s) not in subregions:
        subregions.append((r, s))
    if s not in region_subregions[r]:
        region_subregions[r].append(s)      
```
